```python
import numpy as np
#array indexing and slicing
#1.vectors
arr=np.arange(1,11)
arr
```

```python
arr[6]
```

```python
#slicing-removing a part: default start value=0, default end value=1
n=arr[1:11:2]
n
```

```python
#2.matrices
arr=np.arange(1,31).reshape(6,5)
arr
```

```python
#to print a row, supposedly first
arr[1]
```

```python
#to print an element, supposedly 7
arr[1,1]
```

```python
#to print a column,supposedly first
arr[:,1]
```

```python
#to print 2,3,7,8
arr[0:2,1:3]
```

```python
#3.boolean indexing
arr=np.arange(11,21)
arr
```

```python
#to print even number
bool_idx=arr%2==0
bool_idx
```

```python
#if i want the even values
arr=arr[bool_idx]
arr
```
